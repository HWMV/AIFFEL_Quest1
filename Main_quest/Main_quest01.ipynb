```python
#Main_Quest 1. 입출금 내역 출력
#1. class Account(계좌개설)
#2. 입출금 기록 객체를 모아 리스트화
#3. 리스트(내역)출력
# account_holder = '예금주'
# account_number = '계좌번호'
# balance = '잔액'
# bank_name = '은행명'
class Account:

    account_count = 0 # Q2. 계좌를 세기 위한 클래스 변수 선언
    deposit_count = 0 # Q7.
    account_list = [] # Q8~9. 계좌 저장할 리스트

    def __init__(self, account_holder, balance): # Q1. 생성자, 예금주명, 잔액 받기

        self.account_input_info = input("이름을 입력하세요 : ")
        self.bank_name = "국민은행"
        self.accout_holder = account_holder
        self.balance = balance
        Account.account_list.append(self) # Q8~9.

        # 계좌 번호 랜덤 생성 Q1. (random.randint로 3자리, 2자리, 6자리 생성)
        self.account_number = f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"
        
        # Q2. 계좌 카운트하기(생성될 때마다 1개씩), 클래스 활용 모양
        Account.account_count += 1

        # Q10. 입출력 내역 확인 출력하기
        self.deposit_his = []
        self.withdraw_his = []

        # 계좌 정보 출력
        print("은행명 :", self.bank_name)
        print("예금주 :", self.account_input_info ,"\n"+"계좌번호 :",self.account_number)
        print("잔 액 :", self.balance)

    def get_account_num(self): # Q3. get_account_num() 메소드 만들기

        print(f"총 계좌 개수: {self.account_count}") # 개수 출력하기

    def deposit(self, in_money): # Q4. deposit() 메소드 만들기

        self.balance += int(in_money) # 잔액에 입금액 더하기
        self.deposit_his.append(in_money) # 입금 내역 저장
        
        if in_money < 1: # Q4. 1원 이상만 입금 가능 조건 달기
          print("1원 이상부터 입금 가능 합니다.")
        else:
          print(in_money,"원 입금 됐습니다.","잔고:",self.balance,"원")

        self.deposit_count += 1 # Q7 입금 횟수 세기 위해 작성

        interest = int(self.balance)+(int(self.balance)*0.01)

        if self.deposit_count % 5 == 0: # Q7. 입금 회수가 5의 배수라면 interest 실행해라
          print("이자",interest,"원이 입금 되었습니다.","현재 잔고:",self.balance,"원")
          return

        

    def withdraw(self, out_money): # Q5. 출금 withdraw() 메소드 만들기

        self.balance -= int(out_money)
        minus = int(self.balance) - int(out_money)
        self.withdraw_his.append(out_money)

        if out_money > self.balance:
          print("잔액 이상으로 출금할 수 없습니다.")
        else:
          print(out_money,"원 출금 됐습니다.","잔고:",minus,"원") # minus 변수는 그냥 만들어 본 것

    def display_info(self): # Q6. 정보 보여주는 display_info() method 

        print(" 은행이름:",self.bank_name,"\n","예금주:",self.account_input_info,'\n',"계좌번호:",self.account_number,
              '\n',"잔고:",self.balance)
        
    @classmethod # self 말고 cls로 인스턴스화 하지 않고 호출하려고 선언 했습니다
    def balance_choice(cls):
      
        # Q9. 객체 순회하며 잔고가 100만원 이상인 고객 정보 출력
        print("\n100만원 이상의 잔고를 가진 고객 정보:")

        for account in cls.account_list:
           if account.balance >= 1000000:
              account.display_info()
              print("-----------------")

    def deposit_history(self):
        print("입금 내역:")
        for in_money in self.deposit_his:
            print(in_money)

    def withdraw_history(self):
        print("출금 내역:")
        for out_money in self.withdraw_his:
            print(out_money)

# Q8. 객체 3개 생성하고 리스트에 저장하기
a = Account('',10000000)
b = Account('',200000)
c = Account('',100000)

print(Account.account_list)
Account.balance_choice()

# Q10. 입출금 내역 출력 해보기

account = Account('최현우', 1000000)
account.deposit(50000)
account.deposit(2000)
account.deposit(530000)
account.withdraw(200000)
account.deposit_history()
account.withdraw_history()

# a = Account('', 10000000) # 로그 확인
# a.deposit(5000000) # Q4
# a.withdraw(330000)
# a.deposit(1000000) # Q5
# a.deposit(1000000)
# a.deposit(1000000) # 이자 확인 하기
# a.deposit(1000000)
# print()
# a.display_info()# Q6
# print()
# b = Account('',20000) 
# b.deposit(100000)
# b.withdraw(30000)
# Account.get_account_num # Q3 오류 왜..? 안나올까?
# print("계좌 개수 :",Account.account_count, "개") # 계좌 개수 출력해보기
# print(a.account_number) # 두 줄로 랜덤 계좌 생성 확인!

```
