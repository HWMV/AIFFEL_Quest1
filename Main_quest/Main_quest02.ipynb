```python
import csv
import random
```

```python
class Aiffel:
    student_list = []

    last_num = 0 # XX 순서 기준
    num_input = 0 # Q1. 명단에 학번 부여하기
    id_count = 0 # Q2. 학번의 개수 출력 method 만들기
    score_input_total = 0 # 퀘스트 입력 총점을 추적하는 클래스 변수
    score_penalty_total = 0 # 퀘스트 패널티 총점을 추적하는 클래스 변수

  # 이니셜라이저에 4개의 인자 받기
    def __init__(self, grew_name, entrance_year, cardinal_num, course_name):
          self.grew_name = grew_name
          self.cardinal_num = cardinal_num
          self.course_name = course_name
          self.entrance_year = entrance_year
          self.id = self.generate_cardinal_num() # 메모리 주소가 할당되서 선언
          self.score = 0 # 학생별 스코어 초기화
          Aiffel.student_list.append(self)

    # 학번 조건 만들기
    def generate_cardinal_num(self):

        Aiffel.num_input += 1 # Q1. 학번 부여받기
        Aiffel.id_count += 1 # Q2. 학번의 출력개수 세기

        # 연도의 마지막 2자리
        yy = str(self.entrance_year)[-2:]

        # 기수
        nn = str(self.cardinal_num).zfill(2)  # .zfill : str 앞에 0 채워 넣기
        # print(nn)

        # 과정 종류 c 채워 넣기
        if self.course_name == "리서치":
            c = "1"
        elif self.course_name == "코어":
            c = "2"
        else:
            print("리서치, 코어 중 입력 해주세요!")
            return

        # XX : 가나다순 번호 증가 및 저장
        Aiffel.last_num += 1
        xx = str(Aiffel.last_num).zfill(2)  # 2자리로 표시

        return yy + nn + c + xx

    @classmethod # main quest1 에서 배운 class메소드 활용중
    def num_input_method(cls, students, entrance_year, cardinal_num, course_name):
        for student in students:
            students[student] = Aiffel(student, entrance_year,
                                       cardinal_num, course_name).id
        return students
    @classmethod
    def get_id_num(cls): # Q2. get_id_num 추가
      return cls.id_count

    def Quest_score(self):

            score_input = int(input(f"{self.grew_name}의 점수를 입력하세요(1~3): "))
            if 1 <= score_input <= 3:
                self.score += score_input
                Aiffel.score_input_total += score_input # 입력 점수 업데이트
                return self.score
            else:
                print("점수는 +1점부터 +3점까지만 가능합니다.")

    def Quest_penalty(self):

            penalty = int(input(f"{self.grew_name}의 점수를 입력하세요(1~3): "))
            self.score -= penalty
            Aiffel.score_penalty_total += penalty # 패널티 점수 업데이트


    def display_info(self):

        # Aiffel 인스턴스에 저장된 정보를 출력하는 method
        print("이름:", self.grew_name)
        print("학번:", self.id)
        print("과정종류:", self.course_name)
        print("퀘스트 점수 총점:", self.score)
        print("퀘스트 입력 총점:", self.score_input_total)
        print("퀘스트 패널티 총점:", self.score_penalty_total)

    @classmethod
    def remove_student(cls, name):
        # 중도포기자를 명단에서 삭제하는 메서드
        for student in cls.student_list:
            if student.grew_name == name:
                cls.student_list.remove(student)
                print(f"{name}님이 명단에서 삭제되었습니다.")
                break
        else:
            print(f"{name}님은 명단에 없습니다.")

    @classmethod
    def display_students(cls):
        # 모든 학생들의 명단을 출력하는 메서드
        print("현재 명단:")
        for student in cls.student_list:
            print(student.grew_name)

    # classmethod
    # def save_to_csv(cls, student_dict, filename="core_list.csv"):

        # # 학생 정보를 CSV 파일로 저장하는 method..도저히 모르겠습니다..
        # with open(filename, 'w', newline='') as csvfile:
        #     writer = csv.writer(csvfile)
              # CSV 헤더 작성하기
        #     writer.writerow(['이름', '학번', '과정종류', '퀘스트 점수 총점'])
        #
        #      # 학생 정보를 CSV에 쓰기
        #     for student in students:
        #       s = students[student]
        #       writer.writerow([s.grew_name, s.id, s.course_name, s.score])

core_dic = {'김건': 0, '김경민': 0, '김소연': 0, '김영신': 0, '박근수': 0, '박영수': 0,
            '박영준': 0, '서원영': 0, '신지만': 0, '심지안': 0, '이규상': 0, '이윤상': 0,
             '이진영': 0, '임현석': 0, '전요한': 0, '전재영': 0, '조대희':0}

a1 = Aiffel('박영수', 2023, 25, '코어')

a2 = Aiffel('김건', 2023, 25, '코어')
# print(aiffel.generate_cardinal_num())   # 학번 나오나 로그

id_c = Aiffel.num_input_method(core_dic, 2023, 25, "코어") # 명단 학번 나오는지 확인
print(id_c)
print()
print(Aiffel.get_id_num()) # Q2. get_id_num 출력
a2.Quest_score() # Q3.
a2.Quest_penalty() # Q4.
a2.display_info()  # Q5. 정보 출력
print()
a1.Quest_score() # Q3.
a1.Quest_penalty() # Q4.
a1.display_info()  # Q5. 정보 출력
print()
Aiffel.display_students() # 전체 명단 확인
print()
Aiffel.remove_student('김경민') # Q6. 명단에서 삭제
Aiffel.remove_student('김영신') # Q6. 명단에서 삭제
print()
Aiffel.display_students() # 삭제 명단 확인
print()
# Aiffel.save_to_csv(core_dic)  #Q5. 정보를 csv 파일로 저장

# # 모든 학생의 정보를 CSV 파일로 저장
# Aiffel.save_to_csv(core_dic)

class Aiffel_group(Aiffel): # 부모 클래스 상속 받기

    group_name = ['상생', '신뢰', '열정', '이끔', '성장', '가치', '공유', '확산']

    @classmethod
    def random_group(cls, group_size=2): # 2인 그룹핑

        group_ok = [] # 그룹 된 그루의 리스트
        group_no = cls.student_list.copy()
         # 그룹 해야 될 그루

        while len(group_no) > 0:

          grouping = []

          for i in range(group_size):
            if group_no: # 아직 그룹이 안된 사람 있다면
                student = random.choice(group_no)
                grouping.append(student.grew_name)
                group_no.remove(student) # 위에서 랜덤하고 고르고 넣고 지워서 정리
            else:
              break

          group_ok.append(grouping)

        # 그룹 이름고 그룹핑 완료 명단 출력하기
        for group_name, group in zip(cls.group_name, group_ok):
          print(f"{group_name}:{','.join(group)}")

Aiffel_group.random_group(2) # Q7. 2인 그룹핑 # type 또 오류.. classmethod 사용하기
print()
Aiffel_group.random_group(3) # Q7. 3인 그룹핑, 4인도 arg 4 입력

# Q8. 길드 클래스 만들기
# 위에 그루 그룹핑 구조 참고해서 Try
class Aiffel_guild(Aiffel): # 상속 받는 것 잊어서 오류!

    guild_group_name = ['힌튼','알트만','응','하사비스']

    # Q9. 길드 소속 그루 합산 점수 내기
    guild_scores = {name: 0 for name in guild_group_name}
    guild_members = {name: [] for name in guild_group_name}

    @classmethod
    def guild_random_group(cls,group_size=5):

        guild_group_ok = []
        guild_group_no = cls.student_list.copy()

        while len(guild_group_no) > 0:

          guild_grouping = []

          for i in range(group_size):
            if guild_group_no: # 아직 그룹이 안된 사람 있다면
                student = random.choice(guild_group_no)
                guild_grouping.append(student.grew_name)
                guild_group_no.remove(student) # 위에서 랜덤하고 고르고 넣고 지워서 정리
            else:
              break

          guild_group_ok.append(guild_grouping)

        for guild_group_name, group in zip(cls.guild_group_name, guild_group_ok):
          cls.guild_members[guild_group_name] = group

          print(f"{guild_group_name}:{','.join(group)}")

    @classmethod
    def guild_score(cls, guild_name, score):

        # 길드 점수 업데이트 하기
        if guild_name in cls.guild_scores:
            cls.guild_scores[guild_name] += score
        else:
            print(f"{guild_name} 길드명을 참고해 주세요!")
            return

        # 해당 길드의 그루들의 스코어를 개별로 업데이트
        for student_name in cls.guild_members[guild_name]:
            for student in cls.student_list:
                if student.grew_name == student_name:
                    student.score += score
        # 길드의 총점 출력
        print(f"{guild_name} 길드의 현재 점수: {cls.guild_scores[guild_name]}")

print()
Aiffel_guild.guild_random_group(4) # Q8. 길드 클래스(상속) 길드 랜덤 그룹핑
print()
Aiffel_guild.guild_score('응', 20) # 개인 별로 점수를 줘서 합산하는 방법을 못 찾겠다..
```
